```python
from mlagents_envs.environment import UnityEnvironment
unity_env = UnityEnvironment(file_name="./CarRacing")
```

```python
from gym_unity.envs import UnityToGymWrapper
env = UnityToGymWrapper(unity_env)
```

```python
import numpy as np
np.set_printoptions(precision=3, sign=' ', floatmode="fixed", suppress=True, linewidth=100)

action_size = env.action_space.shape
state_size = env.observation_space.shape

for ep in range(1):
    state = env.reset()
    done = False
    step = 0
    while not done and step<1000:
        action = np.random.uniform([-1, -1, -1], [1, 1, 1], size=action_size)
        state, reward, done, _ = env.step(action)
        print(f"Step: {step:5d}, A: {action}, Pos: {state[:3]}, Idle: {state[-1]}")
        env.render()
        step += 1
```

```python
env.close()
```
